The single largets product by units is the TM 195. This unit seems to have consistent customer base that trends a bit younger, middle of the road education, income, fitness, etc. Keeping this as a catch all product is probably the correct path forward for it

The TM 798 has a very distinctive customer base. They tend to in the 25-23 year old range, married men, with higher educations and incomes. This customer base also trends toward more usage of the product than other groups. The TM 498 seems to be loosing older customers when it comes to utilization. The company should investigage as to why.
